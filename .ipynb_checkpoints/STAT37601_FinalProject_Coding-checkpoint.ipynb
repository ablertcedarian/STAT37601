# Code I

## 1.

The network follows a encoder/decoder structure conditioned on the diffusion time-step t and predicts a vector field for each pixel that tells the Gaussian mean where each noisy sample should move to denoise itself. In essence, ScoreNet is a U-Net with activations modulated by an embedding of the time t, which allows the weights to have varying effects depending on their temporal position.

The time embedding uses the GaussianFourierProjection and a linear layer to produce a vector that represents the current noise level. The encoder is then composed of four convolutional layers to extract hierarchical features while injecting modulation from time with Dense(). We then have a decoder with 4 tconv layers with skip connections that carry spatial detail from the encoder and time modulation again appearing before each GroupNorm (this allows the decoder to vary denoising strength based on how much noise there is at step t). The model also uses the swish activation function to keep gradients smooth.

```python
import torch
import torch.nn as nn
```

## 2.

```python
#@title Defining a time-dependent score-based model (double click to expand or collapse)


class GaussianFourierProjection(nn.Module):
  """Gaussian random features for encoding time steps."""
  def __init__(self, embed_dim, scale=30.):
    super().__init__()
    # Randomly sample weights during initialization. These weights are fixed
    # during optimization and are not trainable.
    self.W = nn.Parameter(torch.randn(embed_dim // 2) * scale, requires_grad=False)
  def forward(self, x):
    x_proj = x * self.W * 2 * np.pi
    return torch.cat([torch.sin(x_proj), torch.cos(x_proj)], dim=-1)


class Dense(nn.Module):
  """A fully connected layer that reshapes outputs to feature maps."""
  def __init__(self, input_dim, output_dim):
    super().__init__()
    self.dense = nn.Linear(input_dim, output_dim)
  def forward(self, x):
    return self.dense(x)[..., None, None]


class ScoreNet(nn.Module):
  """A time-dependent score-based model built upon U-Net architecture."""

  def __init__(self, channels=[32, 64, 128, 256], embed_dim=256, group_num=4):
    
    super().__init__()
    
    self.embed = nn.Sequential(GaussianFourierProjection(embed_dim=embed_dim),
         nn.Linear(embed_dim, embed_dim))
   
    self.conv1 = nn.Conv2d(1, channels[0], 3, stride=1, bias=False)
    self.dense1 = Dense(embed_dim, channels[0])
    self.gnorm1 = nn.GroupNorm(group_num, num_channels=channels[0])
    self.conv2 = nn.Conv2d(channels[0], channels[1], 3, stride=2, bias=False)
    self.dense2 = Dense(embed_dim, channels[1])
    self.gnorm2 = nn.GroupNorm(group_num, num_channels=channels[1])
    self.conv3 = nn.Conv2d(channels[1], channels[2], 3, stride=2, bias=False)
    self.dense3 = Dense(embed_dim, channels[2])
    self.gnorm3 = nn.GroupNorm(group_num, num_channels=channels[2])
    self.conv4 = nn.Conv2d(channels[2], channels[3], 3, stride=2, bias=False)
    self.dense4 = Dense(embed_dim, channels[3])
    self.gnorm4 = nn.GroupNorm(group_num, num_channels=channels[3])

    self.tconv4 = nn.ConvTranspose2d(channels[3], channels[2], 3, stride=2, bias=False)
    self.dense5 = Dense(embed_dim, channels[2])
    self.tgnorm4 = nn.GroupNorm(group_num, num_channels=channels[2])
    self.tconv3 = nn.ConvTranspose2d(channels[2] + channels[2], channels[1], 3, stride=2, bias=False, output_padding=1)
    self.dense6 = Dense(embed_dim, channels[1])
    self.tgnorm3 = nn.GroupNorm(group_num, num_channels=channels[1])
    self.tconv2 = nn.ConvTranspose2d(channels[1] + channels[1], channels[0], 3, stride=2, bias=False, output_padding=1)
    self.dense7 = Dense(embed_dim, channels[0])
    self.tgnorm2 = nn.GroupNorm(group_num, num_channels=channels[0])
    self.tconv1 = nn.ConvTranspose2d(channels[0] + channels[0], 1, 3, stride=1)

    self.rho0 = nn.Parameter(torch.tensor(1.0))  # initialise at 1
    self.rho1 = nn.Parameter(torch.tensor(1.0))
    
    # The swish activation function
    self.act = lambda x: x * torch.sigmoid(x)

  def forward(self, x, t):
    # Obtain the Gaussian random feature embedding for t
    embed = self.act(self.embed(t))

    h1 = self.conv1(x) # ...
    h1 += self.dense1(embed) #...
    h1 = self.gnorm1(h1) # ...
    h1 = self.act(h1) # ...
    h2 = self.conv2(h1) # ...
    h2 += self.dense2(embed)
    h2 = self.gnorm2(h2)
    h2 = self.act(h2)
    h3 = self.conv3(h2)
    h3 += self.dense3(embed)
    h3 = self.gnorm3(h3)
    h3 = self.act(h3)
    h4 = self.conv4(h3)
    h4 += self.dense4(embed)
    h4 = self.gnorm4(h4)
    h4 = self.act(h4)

    h = self.tconv4(h4) # ...
    h += self.dense5(embed) # ...
    h = self.tgnorm4(h)
    h = self.act(h)
    h = self.tconv3(torch.cat([h, h3], dim=1)) # ...
    h += self.dense6(embed)
    h = self.tgnorm3(h)
    h = self.act(h)
    h = self.tconv2(torch.cat([h, h2], dim=1))
    h += self.dense7(embed)
    h = self.tgnorm2(h)
    h = self.act(h)
    h = self.tconv1(torch.cat([h, h1], dim=1))

    mu = self.rho0 * (x - self.rho1 * h)
    return mu
```

## 3.

```python
class Diffusion(nn.Module):
    def __init__(self, model, n_steps, device, min_beta, max_beta):
        super().__init__()
        # Store beta, alpha and \bar alpha 
        self.model = model.to(device)
        self.n_steps = n_steps 
        self.device = device
        
        self.betas = torch.linspace(min_beta, max_beta, n_steps, device=device)
        self.alphas = 1.0 - self.betas 
        self.bar_alphas = torch.cumprod(self.alphas, dim=0) 
        
#         self.register_buffer('betas', self.betas)
#         self.register_buffer('alphas', self.alphas)
#         self.register_buffer('bar_alphas', self.bar_alphas)
        self.register_buffer('sqrt_bar_alphas', torch.sqrt(self.bar_alphas))
        self.register_buffer('sqrt_one_minus_bar_alphas', torch.sqrt(1.0-self.bar_alphas))
    
    def forward_process(self, x0, t):
        # Sample x_{t-1}, x_t given x_0
        assert t.dim() == 1 and t.shape[0] == x0.shape[0]

        B = x0.size(0)

        sqrt_ab_t      = self.sqrt_bar_alphas[t].view(B, 1, 1, 1)
        sqrt_1m_ab_t   = self.sqrt_one_minus_bar_alphas[t].view(B,1,1,1)

        eps_t          = torch.randn_like(x0)
        x_t            = sqrt_ab_t * x0 + sqrt_1m_ab_t * eps_t

        # same for t−1 (clamp so t=1 gives t−1 = 0, the clean data step)
        t_minus_1      = (t - 1).clamp(min=0)
        sqrt_ab_tm1    = self.sqrt_bar_alphas[t_minus_1].view(B,1,1,1)
        sqrt_1m_ab_tm1 = self.sqrt_one_minus_bar_alphas[t_minus_1].view(B,1,1,1)

        eps_tm1        = torch.randn_like(x0)
        x_tm1          = sqrt_ab_tm1 * x0 + sqrt_1m_ab_tm1 * eps_tm1

        return x_tm1, x_t
    
#     @torch.no_grad()
    def predict_next(self,
                     x_t: torch.Tensor,
                     t  : torch.Tensor) -> torch.Tensor:
        # reshape/combine batch dims → (B,1) so ScoreNet’s embed works
        t_in = t.float().view(-1, 1).to(x_t.device)

        out = self.model(x_t, t_in)

        mu = out[0] if isinstance(out, (tuple, list)) else out
        return mu

    def sample_timesteps(self, batch_size):
        return torch.randint(
            low=1, high=self.n_steps, size=(batch_size, ),
            device=self.device,
        )
```

## 4.

```python
def ddpm_mu_loss(diffusion: Diffusion,
                 x0: torch.Tensor) -> torch.Tensor:
    B, *_ = x0.shape
    device = x0.device

    t = diffusion.sample_timesteps(B)          # shape (B,)

    x_tm1, x_t = diffusion.forward_process(x0, t)   # each (B,1,H,W)

    mu_pred = diffusion.predict_next(x_t, t)        # (B,1,H,W)

    alpha_t     = diffusion.alphas[t].view(B, 1, 1, 1)   # broadcastable
    denom_const = 2.0 * (1.0 - alpha_t)                  # (B,1,1,1)

    se   = (x_tm1 - mu_pred).pow(2)                      # (B,1,H,W)
    loss = (se / denom_const).mean(dim=(1, 2, 3))        # (B,)

    return loss.mean()    
```

## 5.

The KL term in the equation involves an expectation. Instead of sampling $x_{t-1}$, we can replace it with its conditional mean, which has the same expectation, but will have lower variance by Rao-Blackwell.
We can plug this deterministic target into the loss.

```python
def ddpm_mu_loss2(diffusion: Diffusion,
                 x0: torch.Tensor) -> torch.Tensor:
    B, *_ = x0.shape
    device = x0.device

    t = diffusion.sample_timesteps(B)

    _, x_t = diffusion.forward_process(x0, t)          # (B,1,H,W)

    abar_t    = diffusion.bar_alphas[t].view(B,1,1,1)        # \bar α_t
    abar_tm1  = diffusion.bar_alphas[(t-1).clamp(min=0)].view(B,1,1,1)
    beta_t    = diffusion.betas[t].view(B,1,1,1)
    alpha_t   = diffusion.alphas[t].view(B,1,1,1)

    coef1 = torch.sqrt(abar_tm1) * beta_t / (1. - abar_t)
    coef2 = torch.sqrt(alpha_t)   * (1. - abar_tm1) / (1. - abar_t)
    x_tm1_mean = coef1 * x0 + coef2 * x_t              # μ_q

    mu_pred = diffusion.predict_next(x_t, t)           # (B,1,H,W)

    loss = ((x_tm1_mean - mu_pred).pow(2) /
            (2. * (1. - alpha_t))).mean(dim=(1,2,3))   # per-sample
    return loss.mean()
```

## 6.

```python
import torchvision, random, math, time
from torch.utils.data import DataLoader, Subset, random_split
from torchvision.transforms import Compose, ToTensor, Lambda
import numpy as np

device  = torch.device("cuda" if torch.cuda.is_available() else "cpu")
epochs  = 60
T       = 200
batch   = 100
beta_1, beta_T = 1e-4, 1e-1
```

```python
def norma(x):
    return x*2. - 1
```

```python
mnist_full = torchvision.datasets.MNIST(
    root=".", train=True, download=True,
    transform=Compose([ToTensor(), Lambda(norma)])  # [0,1] → [-1,1]
)
subset_len = 1_000
splits = random.sample(range(len(mnist_full)), 20*subset_len)
train_subsets = [Subset(mnist_full, splits[i*subset_len:(i+1)*subset_len])
                 for i in range(20)]
```

```python
model_a = ScoreNet().to(device)
model_b = ScoreNet().to(device)      # fresh copy for variant (b)

diff_a = Diffusion(model_a, n_steps=T, device=device,
                   min_beta=beta_1, max_beta=beta_T)
diff_b = Diffusion(model_b, n_steps=T, device=device,
                   min_beta=beta_1, max_beta=beta_T)

# group 0 = all but ρ parameters, group 1 = only ρ0,ρ1
def make_optim(model):
    rho_params   = [p for n,p in model.named_parameters() if n.endswith(("rho0","rho1"))]
    main_params  = [p for n,p in model.named_parameters() if not n.endswith(("rho0","rho1"))]
    opt = torch.optim.Adam([
        {"params": main_params, "lr": 1e-2},
        {"params": rho_params , "lr": 2e-1}
    ])
    return opt

opt_a, opt_b = make_optim(model_a), make_optim(model_b)
scheduler_a  = torch.optim.lr_scheduler.StepLR(opt_a, step_size=20*len(train_subsets),
                                               gamma=0.1)
scheduler_b  = torch.optim.lr_scheduler.StepLR(opt_b, step_size=20*len(train_subsets),
                                               gamma=0.1)
```

```python
def loss_baseline(diff, x0):
    """item 4 loss (stochastic x_{t-1})."""
    B = x0.size(0)
    t = diff.sample_timesteps(B)
    x_tm1, x_t = diff.forward_process(x0, t)     # stochastic target
    mu_pred    = diff.predict_next(x_t, t)
    alpha_t    = diff.alphas[t].view(B,1,1,1)
    return ((x_tm1 - mu_pred).pow(2)/(2*(1-alpha_t))).mean()

def loss_lowvar(diff, x0):
    """item 5 variance-reduced loss (analytic mean target)."""
    B = x0.size(0)
    t = diff.sample_timesteps(B)
    _, x_t = diff.forward_process(x0, t)         # only x_t needed
    abar_t   = diff.bar_alphas[t].view(B,1,1,1)
    abar_tm1 = diff.bar_alphas[(t-1).clamp(min=0)].view(B,1,1,1)
    beta_t   = diff.betas[t].view(B,1,1,1)
    alpha_t  = diff.alphas[t].view(B,1,1,1)

    coef1 = torch.sqrt(abar_tm1)*beta_t/(1-abar_t)
    coef2 = torch.sqrt(alpha_t)*(1-abar_tm1)/(1-abar_t)
    x_tm1_mean = coef1*x0 + coef2*x_t

    mu_pred = diff.predict_next(x_t, t)
    return ((x_tm1_mean - mu_pred).pow(2)/(2*(1-alpha_t))).mean()
```

```python
log = []    # store (epoch, loss_a, loss_b)

for epoch in range(1, epochs+1):
    subset_id = (epoch-1) % 20              # cycle through 20 subsets
    train_loader = DataLoader(train_subsets[subset_id],
                              batch_size=batch, shuffle=True,
                              drop_last=True, 
#                               num_workers=2, 
                              pin_memory=True)

    sum_a = sum_b = 0.0
    for X,_ in train_loader:
        X = X.to(device)

        # ------ (a) baseline ------
        opt_a.zero_grad()
        loss_a = loss_baseline(diff_a, X)
        loss_a.backward()
        opt_a.step()
        sum_a += loss_a.item()

        # ------ (b) low-variance ------
        opt_b.zero_grad()
        loss_b = loss_lowvar(diff_b, X)
        loss_b.backward()
        opt_b.step()
        sum_b += loss_b.item()

        scheduler_a.step();  scheduler_b.step()

    n_batches = len(train_loader)
    log.append((epoch, sum_a/n_batches, sum_b/n_batches))
    print(f"epoch {epoch:02d}: "
          f"baseline={sum_a/n_batches:.4f}  low-var={sum_b/n_batches:.4f}")

final_losses = log[-1][1:]
print("\nFinal avg losses  after 60 epochs")
print("{:>12} {:.6f}\n{:>12} {:.6f}"
      .format("baseline", final_losses[0],
              "low-variance", final_losses[1]))
```

## 7.

Based on the above experiment, we choose the lower variance loss.

```python
class ScoreNetTiny(ScoreNet):          # under-parameterised ↔ risk of under-fit
    def __init__(self): super().__init__(channels=[16, 32,  64, 128])

class ScoreNetBase(ScoreNet):          # reference model from the hand-out
    def __init__(self): super().__init__(channels=[32, 64, 128, 256])

class ScoreNetWide(ScoreNet):          # over-parameterised ↔ risk of over-fit
    def __init__(self): super().__init__(channels=[64,128,256,512])

models = {
    "tiny" : ScoreNetTiny().to(device),
    "base" : ScoreNetBase().to(device),
    "wide" : ScoreNetWide().to(device),
}
```

```python
train_set, val_set = random_split(mnist_full, [1_000, len(mnist_full)-1_000],
                                  generator=torch.Generator().manual_seed(42))
train_loader = DataLoader(train_set, batch_size=100, shuffle=True,
                          drop_last=True, pin_memory=True)
val_loader   = DataLoader(val_set , batch_size=256, shuffle=False,
                          pin_memory=True)
```

```python
T           = 200
beta_1      = 1e-4
beta_T      = 1e-1
epochs_cmp  = 10                    
lr_main     = 1e-2
lr_rho      = 2e-1
step_sched  = torch.optim.lr_scheduler.StepLR
gamma_decay = 0.1                    # ×10 decay every 20 epochs below
```

```python
val_history = {name: [] for name in models}

for name, net in models.items():
    diff = Diffusion(net, n_steps=T, device=device,
                     min_beta=beta_1, max_beta=beta_T)
    opt  = make_optim(net)
    sch  = step_sched(opt, step_size=20*len(train_loader), gamma=gamma_decay)

    print(f"\n training {name} ({sum(p.numel() for p in net.parameters()):,} params)")
    for epoch in range(1, epochs_cmp+1):
        net.train()
        for X,_ in train_loader:
            opt.zero_grad()
            loss = ddpm_mu_loss2(diff, X.to(device))
            loss.backward()
            opt.step()
            sch.step()

        # ---------------- validation ----------------
        net.eval()
        with torch.no_grad():
            val_loss = 0.0
            n_val    = 0
            for X,_ in val_loader:
                bs        = X.size(0)
                val_loss += ddpm_mu_loss2(diff, X.to(device)).item() * bs
                n_val    += bs
        val_loss /= n_val
        val_history[name].append(val_loss)
        print(f"  epoch {epoch:02d}  val-loss = {val_loss:8.5f}")
```

```python
best_arch = min(val_history, key=lambda k: val_history[k][-1])
print("\n===========================================================")
print(f"Best architecture after {epochs_cmp} epochs ⇒  {best_arch.upper()}")
print("===========================================================")
```

```python
# @torch.no_grad()
def sample_from_diffusion(diffusion: Diffusion,
                           n_samples: int = 64,
                           img_shape: tuple = (1, 28, 28)):
    """
    Returns a tensor (n_samples, *img_shape) in [-1,1].
    """
    device = diffusion.device
    T      = diffusion.n_steps

    x_t = torch.randn(n_samples, *img_shape, device=device)

    for t in range(T, 0, -1):
        t_batch = torch.full((n_samples,), t, device=device, dtype=torch.long)

        mu = diffusion.predict_next(x_t, t_batch, 
#                                     no_grad=True
                                   )

        alpha_t = diffusion.alphas[t-1] if t > 0 else diffusion.alphas[0]
        sigma_t = torch.sqrt(1.0 - alpha_t).to(device)

        z = torch.randn_like(x_t) if t > 1 else 0.0    # no noise at last step
        x_t = mu + sigma_t * z

    return x_t.clamp(-1, 1)        # keep the pixel range consistent
```

## 9.

```python
from torchvision.utils import make_grid
```

```python
def sqrtm_psd(A, tol=1e-10):
    """
    Returns the symmetric square-root of a symmetric PSD matrix.
    """
    eigvals, eigvecs = np.linalg.eigh(A)
    A = np.clip(eigvals, a_min=0.0, a_max=None)
    A[A < tol * A.max()] = 0.0
    sqrt_A = np.sqrt(A)
    return (eigvecs * sqrt_A) @ eigvecs.T 

def frechet_Distance(features1, features2, eps=1e-6):
    """
    Compute the FID between two collections of feature vectors 
    """
    mu1 = features1.mean(axis=0)
    mu2 = features2.mean(axis=0)
    diff = mu1 - mu2 
    diff_sq = diff @ diff # |mu_1 - mu_2|^2 
    
    sigma_1 = np.cov(features1, rowvar=False) + eps * np.eye(features1.shape[1])
    sigma_2 = np.cov(features2, rowvar=False) + eps * np.eye(features2.shape[1])
    
    covmean_sqrt = sqrtm_psd(sigma_1 @ sigma_2)
    trace_term = np.trace(sigma_1) + np.trace(sigma_2) - 2 * np.trace(covmean_sqrt)
    return float(diff_sq + trace_term)
```

```python
class MiniInception(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 32, 3, 2, 1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, 2, 1), nn.ReLU(),
            nn.AdaptiveAvgPool2d(1),
        )
    def forward(self, x):
        return self.features(x).view(x.size(0), -1)
    
def get_features(imgs: torch.Tensor, batch=1) -> np.ndarray:
    _feat_extractor = MiniInception().to(device).eval()
    with torch.no_grad():
        feats = []
        for i in range(0, imgs.size(0), batch):
            x = imgs[i:i+batch].to(device)
            feats.append(_feat_extractor(x).cpu())         # CPU → free GPU mem
        return torch.cat(feats, dim=0).numpy().astype(np.float64)
    
def fid_from_feats(real: np.ndarray, fake: np.ndarray) -> float:
    mu1, mu2 = real.mean(0), fake.mean(0)
    sig1 = np.cov(real, rowvar=False)
    sig2 = np.cov(fake, rowvar=False)
    diff = mu1 - mu2
    covmean = sqrtm_psd(sig1 @ sig2)
    return float(diff @ diff + np.trace(sig1) + np.trace(sig2) - 2 * np.trace(covmean))
```

```python
def fid_and_preview(diffusion,
                    test_loader,
                    n_gen: int = 1_000,
                    n_show: int = 20,
                    batch_size: int = 100):
    device = diffusion.device

    real_feats = []
    with torch.no_grad():
        for x, _ in test_loader:
            real_feats.append(get_features(x.to(device)).cpu().numpy())
    real_feats = np.concatenate(real_feats, axis=0).astype(np.float64)

    fake_imgs = []
    with torch.no_grad():
        while len(fake_imgs) < n_gen:
            k = min(batch_size, n_gen - len(fake_imgs))
            fake_imgs.append(sample_from_diffusion(diffusion,
                                                   n_samples=k))
    fake_imgs  = torch.cat(fake_imgs, 0)                      # (n_gen,1,28,28)
    fake_feats = get_features(fake_imgs).cpu().numpy().astype(np.float64)

    fid = fid_from_feats(real_feats, fake_feats)
    print(f"FID ({n_gen} samples): {fid:.4f}")

    grid = make_grid(fake_imgs[:n_show],
                     nrow=min(10, n_show),
                     normalize=True, value_range=(-1, 1))
    plt.figure(figsize=(2*n_show//10, 2))
#     plt.figure()
    plt.axis("off")
    plt.title(f"{n_show} generated samples  –  FID {fid:.3f}")
    plt.imshow(grid.permute(1, 2, 0).cpu())
    plt.show()

    return fid
```

```python
best_net = models["tiny"]      

diffusion_tiny = Diffusion(best_net,
                           n_steps=T,
                           device=device,
                           min_beta=1e-4,
                           max_beta=0.1)
val_history = {name: [] for name in models}

opt  = make_optim(best_net)
sch  = step_sched(opt, step_size=20*len(train_loader), gamma=gamma_decay)

for epoch in range(1, epochs_cmp+1):
    best_net.train()
    for X,_ in train_loader:
        opt.zero_grad()
        loss = ddpm_mu_loss2(diffusion_tiny, X.to(device))
        loss.backward()
        opt.step()
        sch.step()

    best_net.eval()
    with torch.no_grad():
        val_loss = 0.0
        n_val    = 0
        for X,_ in val_loader:
            bs        = X.size(0)
            val_loss += ddpm_mu_loss2(diffusion_tiny, X.to(device)).item() * bs
            n_val    += bs
    val_loss /= n_val
    val_history['tiny'].append(val_loss)
    print(f"  epoch {epoch:02d}  val-loss = {val_loss:8.5f}")
```

```python
fid_score = fid_and_preview(diffusion_tiny,
                            val_loader,
                            n_gen=1000,   # FID on 1 000 samples
                            n_show=20)    # show 20 of them
```

```python
class GaussianFourierProjection(nn.Module):
  """Gaussian random features for encoding time steps."""
  def __init__(self, embed_dim, scale=30.):
    super().__init__()
    # Randomly sample weights during initialization. These weights are fixed
    # during optimization and are not trainable.
    self.W = nn.Parameter(torch.randn(embed_dim // 2) * scale, requires_grad=False)
  def forward(self, x):
    x_proj = x * self.W * 2 * np.pi
    return torch.cat([torch.sin(x_proj), torch.cos(x_proj)], dim=-1)


class Dense(nn.Module):
  """A fully connected layer that reshapes outputs to feature maps."""
  def __init__(self, input_dim, output_dim):
    super().__init__()
    self.dense = nn.Linear(input_dim, output_dim)
  def forward(self, x):
    return self.dense(x)[..., None, None]


class ScoreNet(nn.Module):
  """A time-dependent score-based model built upon U-Net architecture."""

  def __init__(self, channels=[32, 64, 128, 256], embed_dim=256, group_num=4):
    
    super().__init__()
    
    self.embed = nn.Sequential(GaussianFourierProjection(embed_dim=embed_dim),
         nn.Linear(embed_dim, embed_dim))
   
    self.conv1 = nn.Conv2d(1, channels[0], 3, stride=1, bias=False)
    self.dense1 = Dense(embed_dim, channels[0])
    self.gnorm1 = nn.GroupNorm(group_num, num_channels=channels[0])
    self.conv2 = nn.Conv2d(channels[0], channels[1], 3, stride=2, bias=False)
    self.dense2 = Dense(embed_dim, channels[1])
    self.gnorm2 = nn.GroupNorm(group_num, num_channels=channels[1])
    self.conv3 = nn.Conv2d(channels[1], channels[2], 3, stride=2, bias=False)
    self.dense3 = Dense(embed_dim, channels[2])
    self.gnorm3 = nn.GroupNorm(group_num, num_channels=channels[2])
    self.conv4 = nn.Conv2d(channels[2], channels[3], 3, stride=2, bias=False)
    self.dense4 = Dense(embed_dim, channels[3])
    self.gnorm4 = nn.GroupNorm(group_num, num_channels=channels[3])

    self.tconv4 = nn.ConvTranspose2d(channels[3], channels[2], 3, stride=2, bias=False)
    self.dense5 = Dense(embed_dim, channels[2])
    self.tgnorm4 = nn.GroupNorm(group_num, num_channels=channels[2])
    self.tconv3 = nn.ConvTranspose2d(channels[2] + channels[2], channels[1], 3, stride=2, bias=False, output_padding=1)
    self.dense6 = Dense(embed_dim, channels[1])
    self.tgnorm3 = nn.GroupNorm(group_num, num_channels=channels[1])
    self.tconv2 = nn.ConvTranspose2d(channels[1] + channels[1], channels[0], 3, stride=2, bias=False, output_padding=1)
    self.dense7 = Dense(embed_dim, channels[0])
    self.tgnorm2 = nn.GroupNorm(group_num, num_channels=channels[0])
    self.tconv1 = nn.ConvTranspose2d(channels[0] + channels[0], 1, 3, stride=1)

    # The swish activation function
    self.act = lambda x: x * torch.sigmoid(x)

  def forward(self, x, t):
    # Obtain the Gaussian random feature embedding for t
    embed = self.act(self.embed(t))

    h1 = self.conv1(x) # ...
    h1 += self.dense1(embed) #...
    h1 = self.gnorm1(h1) # ...
    h1 = self.act(h1) # ...
    h2 = self.conv2(h1) # ...
    h2 += self.dense2(embed)
    h2 = self.gnorm2(h2)
    h2 = self.act(h2)
    h3 = self.conv3(h2)
    h3 += self.dense3(embed)
    h3 = self.gnorm3(h3)
    h3 = self.act(h3)
    h4 = self.conv4(h3)
    h4 += self.dense4(embed)
    h4 = self.gnorm4(h4)
    h4 = self.act(h4)

    h = self.tconv4(h4) # ...
    h += self.dense5(embed) # ...
    h = self.tgnorm4(h)
    h = self.act(h)
    h = self.tconv3(torch.cat([h, h3], dim=1)) # ...
    h += self.dense6(embed)
    h = self.tgnorm3(h)
    h = self.act(h)
    h = self.tconv2(torch.cat([h, h2], dim=1))
    h += self.dense7(embed)
    h = self.tgnorm2(h)
    h = self.act(h)
    h = self.tconv1(torch.cat([h, h1], dim=1))

    return h
```

```python
class Diffusion:
    def __init__(self, n_steps, min_beta=1e-4, max_beta=1e-1, device="cpu"):
        self.device   = device
        self.n_steps  = n_steps
        self.betas    = torch.linspace(min_beta, max_beta, n_steps, device=device)
        self.alphas   = 1. - self.betas
        self.bar_alph = torch.cumprod(self.alphas, dim=0)
        self.sqrt_bar = torch.sqrt(self.bar_alph)
        self.sqrt_1m  = torch.sqrt(1. - self.bar_alph)

    def timesteps(self, B): return torch.randint(0, self.n_steps, (B,), device=self.device)
```

```python
def eps_loss(model: ScoreNet, diff: Diffusion, x0: torch.Tensor):
    B = x0.size(0)
    t = diff.timesteps(B)
    eps = torch.randn_like(x0)                        
    sqrt_bar = diff.sqrt_bar[t].view(B,1,1,1)
    sqrt_1m  = diff.sqrt_1m [t].view(B,1,1,1)
    x_t      = sqrt_bar * x0 + sqrt_1m * eps            # forward process

    eps_hat  = model(x_t, t.float().view(-1,1))

    # weighting factor w(t) = ((1-α_t)^2)/(α_t(1-ᾱ_t))
    alpha_t = diff.alphas[t].view(B,1,1,1)
    abar_t  = diff.bar_alph[t].view(B,1,1,1)
    w = ((1-alpha_t)**2) / (alpha_t * (1-abar_t))

    return (w * (eps - eps_hat).pow(2)).mean()          # Eq. (6)
```

```python
T          = 200
model      = ScoreNet().to(device)
diff       = Diffusion(T, 1e-4, 1e-1, device)
optimiser  = torch.optim.Adam(model.parameters(), lr=1e-3)
scheduler  = torch.optim.lr_scheduler.StepLR(optimiser, step_size=5000, gamma=0.5)

n_epochs   = 20
for epoch in range(1, n_epochs+1):
    # ---- training ----
    model.train();  running = 0
    for x,_ in train_loader:
        x = x.to(device)
        optimiser.zero_grad()
        loss = eps_loss(model, diff, x)
        loss.backward();  optimiser.step();  scheduler.step()
        running += loss.item()
    train_loss = running/len(train_loader)

    # ---- validation ----
    model.eval();  running = 0
    with torch.no_grad():
        for x,_ in val_loader:
            running += eps_loss(model, diff, x.to(device)).item()
    val_loss = running/len(val_loader)

    print(f"epoch {epoch:02d} ─ train {train_loss:8.5f}   val {val_loss:8.5f}")
```

```python
@torch.no_grad()
def sample_eq5(model:ScoreNet, diff:Diffusion, T, n=64, shape=(1,28,28)):
    model.eval()
    x = torch.randn(n,*shape,device=diff.device)      # x_T ~ N(0,I)
    for t in range(T-1,-1,-1):
        t_b = torch.full((n,),t,device=diff.device,dtype=torch.long)
        eps_hat = model(x, t_b.float().view(-1,1))    # ε̂θ(x_t,t)
        alpha_t = diff.alphas[t];  abar_t = diff.bar_alph[t]
        mu = (1/alpha_t.sqrt())*( x - ((1-alpha_t)/((1-abar_t).sqrt()))*eps_hat )
        if t>0:
            sigma = diff.betas[t].sqrt()
            x = mu + sigma*torch.randn_like(x)
        else:
            x = mu                                   # x_0
    return x.clamp(-1,1)
```
