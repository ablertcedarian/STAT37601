```python
from zipfile import ZipFile
def prep_vocab_emb():

  vocab='word2idx_small'
  with open(datadir+vocab+'.json', 'r') as f:
      w2i = json.load(f)
      num_words = len(w2i.keys())
      print('NUM WORDS',num_words)
  # load in pre-trained word embeddings and  find those from loaded vocab
  word2vec = {}
  start = time.time()
  with ZipFile(datadir+'glove.6B.50d.txt.zip', 'r') as ZZ:
    for filename in ZZ.namelist():
       with ZZ.open(filename, 'r') as f:
        for i,line in enumerate(f):
          values = line.split()
          word = values[0]
          vec = np.asarray(values[1:], dtype='float32')
          word2vec[word] = vec

  # prepare embedding matrix
  WordEmbeddings = np.zeros((num_words+1, 50)) # Index 0 will be zero
  for word, i in w2i.items():
      if word in word2vec:
          WordEmbeddings[i] = word2vec[word]

  return WordEmbeddings, w2i
```

```python

# Remove puntuation
def remove_punctuation(s):
    return s.translate(str.maketrans('','',string.punctuation+"\n"))

# Convert sentence to word list
def ConvertSentence2Word(s):
    return(word_tokenize(remove_punctuation(s).lower()))

# Convert sentence to word index list
def ConvertSent2Idx(s):
    s_temp = [w for w in ConvertSentence2Word(s) if w in w2i]
    temp = []
    for w in s_temp:
        temp.append(w2i[w])
    return(temp)

# Divide Document into  sentences and process each sentence
def ConvertDoc2List(doc):
    temp_doc = sent_tokenize(doc)
    temp = []
    for i in range(len(temp_doc)):
        if(len(ConvertSent2Idx(temp_doc[i]))>=1):      # Prevent empty sentence
            temp.append(ConvertSent2Idx(temp_doc[i]))
    return(temp)

# Convert List type of document to array of fixed (length of max sentence). Pad shorter sentenves
# with 0's
def ConvertList2Array(docs):
    ms = len(docs)
    mw = len(max(docs, key=len))
    result = np.zeros((ms,mw))
    for i, line in enumerate(docs):
        for j, word in enumerate(line):
            result[i,j] = word
    return result
```

```python
def data_to_array(X_t,Y_t):

    X_t_data=[]
    Y_t_data=[]
    p=len(w2i.keys())
    for i in range(len(X_t)):
        X_input = ConvertDoc2List(X_t[i])
        if(len(X_input)<1):
            continue
        X_input= torch.LongTensor(ConvertList2Array(X_input))
        Y_t_data.append(Y_t[i])
        X_t_data.append(X_input.to(device))
    Y_t_data=torch.tensor(np.array(Y_t_data).reshape((-1,1))).type(torch.long).to(device)
    return X_t_data, Y_t_data
```

```python
def load_data(num,corpus):

  if corpus=='beer':
    with open(datadir + 'labeled.json', 'r') as fopen:
        brv=json.load(fopen)

    X=[]
    Y=[]
    for i,b in enumerate(brv):
        if (i<num):
            X.append(b['review'])
            v=b['overall']
            y=0
            if v>=14:
                y=1

            Y.append(y)
    del brv

  else:
    npz = np.load(datadir + 'yelp_review_small.npz', allow_pickle=True)
    data = npz['arr_0']
    X = data[:,0]  # Text
    Y = data[:,1]  # Label
    Y=Y-1
    del data
  return X, Y
```

```python
def get_data(X,Y):
    X,Y=data_to_array(X,Y)
    ii = np.int64(np.arange(0, len(X), 1))
    np.random.shuffle(ii)
    XX= [X[i] for i in ii]
    X=XX
    # X=X[ii]
    Y=Y[ii]
    num = len(X)
    nntr = np.int32(.8 * num)
    nnva = np.int32(.82 * num)
    X_train_data = X[0:nntr]

    y_train_data = Y[0:nntr]
    X_val_data = X[nntr:nnva]

    y_val_data = Y[nntr:nnva]
    X_test_data = X[nnva:num]

    y_test_data = Y[nnva:num]



    return X_train_data,  y_train_data, X_val_data,  y_val_data, X_test_data, y_test_data
```

# 1.2 b)

The version in class applies an additional convolutional and relu projection, which implies that it additionally slides a kernel and therefore touches more positions than a simple point-wise linear map. In essence, we're building more local interactions before global attention. The version here is simpler and doesn't have that.

# 1.2 c)

In the simpler version of the self attention layer we have here, there is no positional information encoded, which means it doesn't matter if we permute the words in each sentence. 

```python
# Define the class of Target Attention.
class TargetAttention(Module):
    def __init__(self, input_dim,dropout_rate):
        super(TargetAttention, self).__init__()
        # define the transfer embedding vector T\in R^{1*d}
        self.target = nn.Parameter(torch.empty((1,input_dim)))
        nn.init.kaiming_uniform_(self.target)
        self.input_dim=input_dim
        self.sq_input_dim = np.sqrt(input_dim)
        self.dropout = nn.Dropout(dropout_rate)

    #define the target attention process: softmax(T*V^T/sqrt(d))*V
    def target_att(self, t, k, v):

        # Dimensions: batch_size x 1 x number_of_tokens
        out=torch.matmul(t, k.permute(0,2,1))/self.sq_input_dim
        sf=torch.softmax(out,2)

        # Dimensions: batch_size x 1 x embedd_dim
        targ=torch.matmul(sf,v)

        # Dimensions: batch_size x embedd_dim x 1
        targ= targ.permute(0, 2, 1)
        return targ

    def forward(self, inputk, input):
        batch_size = input.size(0)

        # Make the target parameter have the same nunber of dimensions as the inputk, input.
        output = self.target_att(self.target.expand(batch_size, 1, self.input_dim), inputk, input)

        return output
```

# 1.2 d)

```python
class SelfAttention(nn.Module):
    """
    Single-head self-attention.
    • Linear mode   : Q,K,V are simple affine maps of each token (your original code)
    • Convolutional : Q,K,V = ReLU(Conv1d(x, …)), which lets each position
                      look at its neighbours before the attention step.
    Args
    ----
    d_model      : embedding / hidden size
    use_conv     : if True use Conv1d+ReLU projections, else nn.Linear
    kernel_size  : width of the 1-D convolution (only matters when use_conv=True)
    """
    def __init__(self,
                 d_model: int,
                 use_conv: bool = False,
                 kernel_size: int = 3):
        super().__init__()
        self.d_model  = d_model
        self.use_conv = use_conv

        if use_conv:
            # Conv1d expects [B, C, L].  We keep out_channels = d_model
            padding = kernel_size // 2      # keep sequence length unchanged
            self.q_proj = nn.Sequential(
                nn.Conv1d(d_model, d_model,
                          kernel_size=kernel_size,
                          padding=padding,
                          bias=True),
                nn.ReLU(inplace=True)
            )
            self.k_proj = nn.Sequential(
                nn.Conv1d(d_model, d_model,
                          kernel_size=kernel_size,
                          padding=padding,
                          bias=True),
                nn.ReLU(inplace=True)
            )
            self.v_proj = nn.Sequential(
                nn.Conv1d(d_model, d_model,
                          kernel_size=kernel_size,
                          padding=padding,
                          bias=True),
                nn.ReLU(inplace=True)
            )
        else:
            # Original point-wise linear projections
            self.q_proj = nn.Linear(d_model, d_model)
            self.k_proj = nn.Linear(d_model, d_model)
            self.v_proj = nn.Linear(d_model, d_model)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """
        x : (batch, seq_len, d_model)
        Returns
        -------
        out : (batch, seq_len, d_model)
        """
        if self.use_conv:
            # ── Conv1d expects channels before length ───────────────────────────
            x_t = x.transpose(1, 2)          # (B, d_model, L)

            q = self.q_proj(x_t).transpose(1, 2)  # back to (B, L, d_model)
            k = self.k_proj(x_t).transpose(1, 2)
            v = self.v_proj(x_t).transpose(1, 2)
        else:
            q = self.q_proj(x)               # (B, L, d_model)
            k = self.k_proj(x)
            v = self.v_proj(x)

        # ── scaled dot-product attention ───────────────────────────────────────
        scores = torch.matmul(q, k.transpose(-2, -1))          # (B, L, L)
        scores = scores / math.sqrt(self.d_model)
        attn   = F.softmax(scores, dim=-1)

        out = torch.matmul(attn, v)                            # (B, L, d_model)
        return out
```

# 1.2 e)

This version of the target attention layer doesn't apply conv1d and then elu projections to the key and value weights as the version discussed in class did.

```python
## Define the class of convolutional processing of input
class CELL(nn.Module):
    def __init__(self, input_dim, kernel_dim, dropout_rate, mtype): #convc_cnt=1
        super(CELL, self).__init__()
        self.input_dim = input_dim
        self.cell = SelfAttention(input_dim, use_conv=mtype)
        self.dropout = nn.Dropout(dropout_rate)

    def forward(self, input):
        x=input
        hidden= F.relu(self.dropout(self.cell(x))+x)
        output = F.layer_norm(hidden, hidden.size()[1:])

        return output
```

```python
class Hierarchical(nn.Module):
    def __init__(self, num_emb, input_dim, dropout_rate, pretrained_weight, untrained_weight):
        super(Hierarchical, self).__init__()
        # Define the initialization of embedding matrix
        # One is using the pretrained weight matrix, the other is initialized randomly.
        self.id2vec_pretrained = nn.Embedding(num_emb, input_dim)
        self.id2vec_pretrained.weight.data.copy_(pretrained_weight)
        self.id2vec_pretrained.weight.requires_grad=False
        self.id2vec_untrained = nn.Embedding(num_emb, input_dim)
        self.id2vec_untrained.weight.data.copy_(untrained_weight)
        self.id2vec_untrained.requires_grad = True
        self.dropout = nn.Dropout(dropout_rate)

        self.loss=nn.CrossEntropyLoss()

    def accuracy(self,y_pred,y):
        # compute the number of correct of the prediction.
        _, pred=y_pred.max(1)
        acc=pred.eq(y)

        return acc
```

```python
class HSAN(Hierarchical):
    def __init__(self, input_dim, num_class, kernel_dim,
                  dropout_rate, num_emb, pretrained_weight, untrained_weight, ltype, mtype):
        super(HSAN, self).__init__(num_emb, input_dim, dropout_rate, pretrained_weight, untrained_weight)
        self.saw = CELL(input_dim, kernel_dim, dropout_rate, mtype)
        self.sas = CELL(input_dim, kernel_dim, dropout_rate, mtype)
        self.ltype=ltype
        self.taw = TargetAttention(input_dim, dropout_rate)
        self.tas = TargetAttention(input_dim, dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.cls = nn.Linear(input_dim, num_class)
        nn.init.xavier_normal_(self.cls.weight)

    def predict(self, x):

        # Get embedding produce an array of number_of_sentences x number_of_words x embedd_dim.
#         input = self.id2vec_untrained(x)
        input = self.id2vec_pretrained(x)

        # Word level self-attention
        # Out Dim: number_of_sentences x number_of_words x embedd_dim
        hiddenw=self.saw(input)

        # Out Dim: number_of_sentences x embedd_dim x 1
        hiddenw = self.taw(hiddenw,hiddenw)

        # Out Dim: 1 x number_of_sentences x embedd_dim (batch_size = 1)
        hiddenw = hiddenw.permute(2, 0, 1)

        # Sentence level self-attention
        # Out Dim: 1 x number_of _sentences x embedd_dim
        hiddens=self.sas(hiddenw)

        # Out Dim: 1 x embedd_dim  x 1
        hiddens = self.tas(hiddens,hiddens)

        # Get logits
        logits = self.cls(hiddens.squeeze(-1))

        return logits

    def forward(self, x,  y):
        logits = self.predict(x)
        if logits.shape[1]==1:
            logits=torch.cat((-logits,logits),dim=1)
        loss = self.loss(logits, y) #
        accuracy = self.accuracy(logits, y) #
        return loss, accuracy
```

```python
def val_score(model, data, labels):
    #Test the model accuracy for the test data set (or valid data set)
    correct = 0.
    #l = labels.shape[1]

    for i in range(len(data)):
        vloss, vacc = model.forward(data[i], labels[i])
        correct = correct + vacc # Compute the total correct number
    val_acc = correct/len(data)

    return val_acc
```

```python
def run_epochs(model,X_train_data, y_train_data, optimizer,step=0):

  bestscore = 0.
  seed=14543
  torch.manual_seed(seed)
  np.random.seed(seed)
  print("Traing data size:",len(y_train_data))

  n_epochs = 5 # When we remove the dropout layer and softmax layer, it converges fast.
  model.train()
  train_acc=[]
  valid_acc=[]
  # Then we run the model in each epochs.
  num_train=len(X_train_data)
  ii=list(range(num_train))
  t1=time.time()
  for epoch in range(n_epochs):
      np.random.shuffle(ii)
      accuracy = 0.
      tot_loss=0.
      for i,j in enumerate(ii):

          optimizer.zero_grad()
          loss, cor = model.forward(x=X_train_data[j],y=y_train_data[j])
          loss.backward()
          optimizer.step()
          accuracy = accuracy + cor.item()
          tot_loss+=loss.item()
          if (np.mod(i,5000)==0 or i==num_train-1) and i>0:
            val_sc=val_score(model, X_val_data, y_val_data)
            t2=time.time()
            print("iter %i, loss, %.3f, training accuracy: %.2f, validation accuracy: %.2f, time: %.3f" % (
              i, tot_loss/i,accuracy/i, val_sc,t2-t1))
            t1=t2
      accuracy = accuracy / num_train
      train_acc.append(accuracy)
      # test the model on the valid set
      model = model.eval()
      valscore = val_score(model,X_val_data,y_val_data)
      model = model.train()
      valid_acc.append(valscore)

      # save the best model
      if valscore >= bestscore:
          bestscore = valscore
#           save_path = torch.save(model,datadir+"models/beer_py_small_"+model.ltype+"_step"+str(epoch)+".pkl")
      temptime = datetime.timedelta(seconds=round(time.time() - t1))
      print("epoch %i, training accuracy: %.2f, validation accuracy: %.2f," % (
          epoch + 1, accuracy * 100, valscore * 100), "time: ", temptime)
      accuracy = 0

  totaltime = datetime.timedelta(seconds=round(time.time() - t1))
  testscore = val_score(model, X_test_data, y_test_data)
  print("\ntest accuracy: %.2f" % (testscore*100),"%")
  print("\nTime:", totaltime)
  return model, train_acc, valid_acc
```

```python
def setup_model(WordEmbeddings, mtype=False, ltype="SA"):
    # Set the parameters
    input_dim = WordEmbeddings.shape[1] # = 50
    kernel_dim = 3  #
    num_words = WordEmbeddings.shape[0] # number of words in the embedding matrix
    dropout_rate = 0.1
    pretrained_weight = torch.Tensor(WordEmbeddings) # Word Embedding matrix 1
    untrained_weight = torch.Tensor(WordEmbeddings.shape[0],WordEmbeddings.shape[1])
    untrained_weight = nn.init.xavier_normal_(untrained_weight) # Word Embedding matrix 2
    num_class = 2

    # Define the model
    model = HSAN(input_dim, num_class, kernel_dim,
                dropout_rate, num_words, pretrained_weight, untrained_weight, ltype, mtype)
    optimizer=torch.optim.Adam(model.parameters()) # define the optimizer function
    num=0
    for k,p in model.named_parameters():
       if p.requires_grad:
         print(k,p.shape)
         num+=p.numel()
    print('number of paramters',num)
    # If we use gpu to run the model.
    model = model.to(device)
    return model, optimizer
```

# 2b)

```python
import time, numpy as np, pandas as pd, scipy.sparse as sp
from sklearn.linear_model import LogisticRegression
from sklearn.svm       import LinearSVC
from sklearn.metrics   import accuracy_score
```

```python
X_trainval = sp.vstack([X_train, X_val])
y_trainval = np.concatenate([y_train, y_val])

C_grid = [1e-3, 1e-2, 1e-1, 1, 10, 100]
```

```python
def best_model(est_cls, fit_kwargs):
    best, best_C, best_val, best_time = None, None, -np.inf, None
    for C in C_grid:
        clf = est_cls(C=C, **fit_kwargs)
        t0  = time.perf_counter()
        clf.fit(X_train, y_train)
        dt  = time.perf_counter() - t0
        val_acc = accuracy_score(y_val, clf.predict(X_val))
        if val_acc > best_val:
            best, best_C, best_val, best_time = clf, C, val_acc, dt
    return best, best_C, best_val, best_time

# Logistic regression (liblinear → works on sparse)
log_best, log_C, log_val_acc, log_fit_t = best_model(
    LogisticRegression,
    dict(penalty="l2", solver="liblinear", max_iter=5000)
)
# Hinge-loss linear SVM
svm_best, svm_C, svm_val_acc, svm_fit_t = best_model(
    LinearSVC,
    dict(loss="hinge", max_iter=50000)
)

# Retrain both on train+val for final test score
t0 = time.perf_counter();  log_best.fit(X_trainval, y_trainval);  log_re_t = time.perf_counter()-t0
t0 = time.perf_counter();  svm_best.fit(X_trainval, y_trainval);  svm_re_t = time.perf_counter()-t0

log_test_acc = accuracy_score(y_test,  log_best.predict(X_test))
svm_test_acc = accuracy_score(y_test,  svm_best.predict(X_test))

out = pd.DataFrame({
    "model"              : ["LogisticRegression", "LinearSVC (hinge)"],
    "best C"             : [log_C,  svm_C],
    "val accuracy"       : [log_val_acc, svm_val_acc],
    "test accuracy"      : [log_test_acc, svm_test_acc],
    "train time (best) s": [log_fit_t,   svm_fit_t],
    "retrain time s"     : [log_re_t,    svm_re_t]
})
```

# 2c)

The results from the logistic classification method are pretty comparable (quite similar test accuracy) to the hierarchical attention method. Notably, it is also much faster to train.
