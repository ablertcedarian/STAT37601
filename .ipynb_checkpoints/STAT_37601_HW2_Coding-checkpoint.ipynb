```python
# Load the data
import json
speeches = []
with open("data/speeches.json") as f:
    for line in f:
        speeches.append(json.loads(line))
```

## a)

```python
import string
import re

def clean_and_split(s):
    # string.punctuation is a string with all the different punctuation symbols.
    # ’tr’ is an object that will assign blank space to each punctuation symbol.
    tr=str.maketrans(string.punctuation," "*len(string.punctuation))
    s=s.lower().translate(tr)
    # replace \r\n
    s = re.sub('(\r\n)+',' ',s)
    # replace whitespace substrings with one whitespace and remove leading/trailing whitespaces
    s = re.sub(' +',' ',s.strip())
    s=s.split(' ')
    return s
```

We do some pre-processing as suggested by cleaning and splitting the text of each speech and then counting the occurrences of each word overall and within each speech.

We then remove the 20 most common words using heapq's nlargest function from our total word bag and the word bag of each speech.

We then remove words that appear fewer than 50 times and words that are numeric strings using a similar pattern.

```python
from collections import defaultdict 

total_appearances = defaultdict(int)
total_speech_appearances = defaultdict(int)
speech_appearances = defaultdict(lambda: defaultdict(int))
for i, speech_json in enumerate(speeches):
    text = speech_json['text']
    cleaned_text = clean_and_split(text)
    
    seen = {}
    for word in cleaned_text:
        total_appearances[word] += 1
        
        if word not in seen:
            total_speech_appearances[word] += 1
            seen[word] = True
        
        speech_appearances[i][word] += 1

print(len(total_appearances))
# Remove the 20 most common words
from heapq import nlargest

most_common = nlargest(20, total_appearances.items(), key=lambda x: x[1])
for (most_common_word, count) in most_common:
    del total_appearances[most_common_word]
    del total_speech_appearances[most_common_word]
    for key, speech_appearance in speech_appearances.items():
        if speech_appearance.get(most_common_word, 0) > 0:
            del speech_appearance[most_common_word]

print(len(total_appearances))
# Remove words that appear fewer than 50 times and numbers
for (word, count) in list(total_appearances.items()):
    if count < 50:
        del total_appearances[word]
        del total_speech_appearances[word]
        for key, speech_appearance in speech_appearances.items():
            if speech_appearance.get(word, 0) > 0:
                del speech_appearance[word]
    elif word.isnumeric():
        del total_appearances[word]
        del total_speech_appearances[word]
        for key, speech_appearance in speech_appearances.items():
            if speech_appearance.get(word, 0) > 0:
                del speech_appearance[word]

print(len(total_appearances))
```

Finally, we generate a tf_idf vector for each speech using the given equation by looping over all the speeches and then each word in our total vocabulary. 

```python
import numpy as np

tf_idfs = []
total_speech_count = len(speech_appearances)
for key, speech_appearance in speech_appearances.items():
    tf_idf = []
    for word, total_appearance in total_appearances.items():
        tf_idf.append(speech_appearance.get(word, 0) * np.log(total_speech_count / total_speech_appearances.get(word, 0)))
    tf_idfs.append(tf_idf)
```

# b)

A function to calculate sine similarity as described in the handout

```python
def calculate_similarity(tf_idf_1, tf_idf_2):
    return np.cos( (np.dot(tf_idf_1, tf_idf_2)) / (np.linalg.norm(tf_idf_1) * np.linalg.norm(tf_idf_2)))
```

We do a nested for-loop (O(n^2)) to calculate the similarity of every pair of speeches and then, depending on the president that gave the speech, save it. In order to save the top 50 pairs for different and the same presidents while not maintaining giant lists, we do a binary search to check if the current pair would fall within the top 50 so far, and insert it using binary search if so. We then truncate the running list to 50 elements.

```python
import bisect 

diff_pres = []
same_pres = []
sim_pres = []

for i, tf_idf_1 in enumerate(tf_idfs):
    i_president = speeches[i]['president']
    
    for j, tf_idf_2 in enumerate(tf_idfs):
        j_president = speeches[j]['president']
        if i <= j:
            continue
        
        similarity = - calculate_similarity(tf_idf_1, tf_idf_2)
        
        if i_president == j_president:
            pos = bisect.bisect(same_pres, (similarity, i, j))
            if pos <= 50:
                bisect.insort_left(same_pres, (similarity, i, j))
            
            same_pres = same_pres[:50]
        else:
            pos = bisect.bisect(diff_pres, (similarity, i, j))
            if pos <= 50:
                bisect.insort_left(diff_pres, (similarity, i, j))
            
            diff_pres = diff_pres[:50]
```

We establish an index of all the speeches keyed by the president that gave the speech. We then use this in a nested for-loop to check all pairs of presidents, calculating the average similarity across each pair of speeches and then following the same insertion logic as above to save the top 25 most similar pairs.

```python
speeches_by_pres = defaultdict(list)

for i, speech in enumerate(speeches):
    speeches_by_pres[speech['president']] += [i]

for i, (i_pres, i_indices) in enumerate(speeches_by_pres.items()):
    for j, (j_pres, j_indices) in enumerate(speeches_by_pres.items()):
        if i <= j:
            continue
        
        sim_scores = []
        for i_i in i_indices:
            for j_j in j_indices:
                sim_scores.append(calculate_similarity(tf_idfs[i_i], tf_idfs[j_j]))
        
        avg_sim_score = - np.mean(sim_scores)
        
        pos = bisect.bisect(sim_pres, (avg_sim_score, i_pres, j_pres))
        if pos <= 25:
            bisect.insort_left(sim_pres, (avg_sim_score, i_pres, j_pres))
        
        sim_pres = sim_pres[:25]
```

The above results (we print out a pair of speeches by different presidents rated as very similar) demonstrate that even speeches from different presidents can appear very similar when measured with simple cosine similarity on tf-idf vectors. This is probably partly because the genre of the State of the Union addresses shares a lot of common language, rhetorical conventions, and boilerplate expressions. Therefore, our model may overemphasize terms that are generic (though we do combat this by removing the 20 most common words) or structurally necessary in such addresses.
A better model of similarity could try to:
- capture semantic context with word embeddings or sentence embeddings to provide a more nuanced understanding of each sentence
- discount common rhetorical language (like "America" or "the people") while emphasizing distinct terms
- capture phrase-level semantics 
- capture variation within a speech

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# A fnction to run k-means with a specified nuber of clusters and print the cluster size 
def run_kmeans(tf_idf_mat, n_clusters):
    model = KMeans(n_clusters=n_clusters, max_iter=50, random_state=42)
    clusters = model.fit_predict(tf_idf_mat)
    print(f"Number of clusters: {n_clusters}")
    # Count the number of speeches in each cluster
    unique, counts = np.unique(clusters, return_counts=True)
    print(dict(zip(unique, counts)))
    return clusters, model

# We loop over different numbers of clusters to experiment
for n in [5, 10, 15]:
    clusters, model = run_kmeans(tf_idf_mat, n)
    # Optionally visualize the distribution of speeches per cluster
    sns.barplot(x=list(np.unique(clusters)), y=list(np.bincount(clusters)))
    plt.title(f"Cluster distribution for {n} clusters")
    plt.xlabel("Cluster label")
    plt.ylabel("Number of speeches")
    plt.show()
```

We observe that across the range of clusters tested, it is generally true that some clusters contain many speeches while others have only a few. This skewed distribution may suggest that some topics or styles are more common, while a few clusters capture more unique speeches. 

Regarding interpretability, since the tf-idf vectors are fundamentally capturing word frequency information, the clusters may be grouping together speeches with similar rhetorical styles or recurring phrases. This does not mean, however, that they will correspond neatly to policy differences or historical contexts, since speeches by different presidents might end up in the same cluster if they use similar language to discuss themes. At best, it seems that tf-idf is able to capture commonalities in word choice, which would imply that some clusters, identified perhaps by specific sets of word choices, contain many speeches, while others contain very few due to their more unique word choices.

## i., ii., iii.

We first compute the distance matrix A and then compute the weight matrix

```python
A = np.zeros((len(tf_idfs), len(tf_idfs)))

for i, i_tf_idf in enumerate(tf_idfs):
    for j, j_tf_idf in enumerate(tf_idfs):
        A[i][j] = np.linalg.norm(np.array(i_tf_idf) - np.array(j_tf_idf))
```

We then calculate W and W_tilde as described in class over a variety of h values by using various coefficients for h=np.mean(A)

```python
for h_coef in [0.3, 0.10, 0.05, 0.025, 0.01, 0.007]:
    h = np.mean(A)
    print(f"\nAnalyzing for h = {h*h_coef:.2f}")
    W = np.exp( -A / (h * h_coef))

    d = np.sum(W, axis=1)
    D_inv_sqrt = np.diag(1 / np.sqrt(d))

    W_tilde = D_inv_sqrt @ W @ D_inv_sqrt

    # i. 
    # We do an eigen-decomposition, sort the eigenvalues and then grab the top 30 for plotting 
    
    # Eigen-decomposition
    eigvals, eigvecs = np.linalg.eigh(W_tilde)
    sorted_idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[sorted_idx]
    eigvecs = eigvecs[:, sorted_idx]

    # Print the top eigenvalue to confirm it's 1
    print("Top eigenvalue:", eigvals[0])

    # Plot the top 30 eigenvalues
    plt.figure(figsize=(8,5))
    plt.plot(eigvals[:30], 'bo-')
    plt.xlabel('Index')
    plt.ylabel('Eigenvalue')
    plt.title('Top 30 Eigenvalues of the Normalized Affinity Matrix L')
    plt.show()
    
    # ii.
    # We take the top 3 eigenvectors, left-multiply by D^{-1/2} and then plot the three of them 
    
    # Project
    v1 = D_inv_sqrt @ eigvecs[:, 0]
    v2 = D_inv_sqrt @ eigvecs[:, 1]
    v3 = D_inv_sqrt @ eigvecs[:, 2]

    # Create a figure with 1 row and 3 columns
    fig, axs = plt.subplots(1, 3, figsize=(15, 4))

    # Plot v1
    axs[0].plot(v1, marker='o', linestyle='-', color='b')
    axs[0].set_title('Vector v1 (eigenvalue 1)')
    axs[0].set_xlabel('Speech index')
    axs[0].set_ylabel('v1 value')
    axs[0].grid(True)

    # Plot v2
    axs[1].plot(v2, marker='o', linestyle='-', color='r')
    axs[1].set_title('Vector v2')
    axs[1].set_xlabel('Speech index')
    axs[1].grid(True)

    # Plot v3
    axs[2].plot(v3, marker='o', linestyle='-', color='g')
    axs[2].set_title('Vector v3')
    axs[2].set_xlabel('Speech index')
    axs[2].grid(True)

    plt.tight_layout()
    plt.show()
    
    # iii.
    # We plot v2 and v3
    
    # Scatter plot of v2 vs v3
    plt.figure(figsize=(6, 6))
    plt.scatter(v2, v3)
    plt.title(f'Scatter plot of v2 vs v3 (h = {h * h_coef:.2f})')
    plt.xlabel('v2')
    plt.ylabel('v3')
    plt.grid(True)
    plt.show()
```

We observe the (v2,v3) scatter plots over a variety of h values. In some cases, we see some clustering in distinct groups (in the h=1.27 case for instance) or more "line" like shapes (like in h=0.89) or even a relatively even scattering (as in h=3.17). These reflect the underlying geometry of the data as encoded by the similarity matrix -- so the distinct groupings and lines we observe reflect clear communities and/or groupings of speeches (potentially by topic or style). Conversely, the lack of structure in some cases reflects that there might not be particularly clear groupings by topic, etc.

## iv.

We gather the years of all the speeches and also normalize the tf_idfs

```python
years = [speech['year'] for speech in speeches]
```

```python
tf_idfs_norm = [tf_idf / np.linalg.norm(tf_idf) for tf_idf in tf_idfs]
```

We then repeat the above process from i. - iii. on this new st of tf_idfs

```python
A = np.zeros((len(tf_idfs), len(tf_idfs)))

for i, i_tf_idf in enumerate(tf_idfs_norm):
    for j, j_tf_idf in enumerate(tf_idfs_norm):
        A[i][j] = np.linalg.norm(np.array(i_tf_idf) - np.array(j_tf_idf))
for h_coef in [0.3, 0.10, 0.05, 0.025, 0.01, 0.007]:
    h = np.mean(A)
    print(f"\nAnalyzing for h = {h*h_coef:.2f}")
    W = np.exp( -A / (h * h_coef))

    d = np.sum(W, axis=1)
    D_inv_sqrt = np.diag(1 / np.sqrt(d))

    W_tilde = D_inv_sqrt @ W @ D_inv_sqrt

    # Eigen-decomposition
    eigvals, eigvecs = np.linalg.eigh(W_tilde)
    sorted_idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[sorted_idx]
    eigvecs = eigvecs[:, sorted_idx]

    # Print the top eigenvalue to confirm it's 1
    print("Top eigenvalue:", eigvals[0])

    # Plot the top 30 eigenvalues
    plt.figure(figsize=(8,5))
    plt.plot(eigvals[:30], 'bo-')
    plt.xlabel('Index')
    plt.ylabel('Eigenvalue')
    plt.title('Top 30 Eigenvalues of the Normalized Affinity Matrix L')
    plt.show()
    
    # Project
    v1 = D_inv_sqrt @ eigvecs[:, 0]
    v2 = D_inv_sqrt @ eigvecs[:, 1]
    v3 = D_inv_sqrt @ eigvecs[:, 2]

    # Create a figure with 1 row and 3 columns
    fig, axs = plt.subplots(1, 3, figsize=(15, 4))

    # Plot v1
    axs[0].plot(v1, marker='o', linestyle='-', color='b')
    axs[0].set_title('Vector v1 (eigenvalue 1)')
    axs[0].set_xlabel('Speech index')
    axs[0].set_ylabel('v1 value')
    axs[0].grid(True)

    # Plot v2
    axs[1].plot(v2, marker='o', linestyle='-', color='r')
    axs[1].set_title('Vector v2')
    axs[1].set_xlabel('Speech index')
    axs[1].grid(True)

    # Plot v3
    axs[2].plot(v3, marker='o', linestyle='-', color='g')
    axs[2].set_title('Vector v3')
    axs[2].set_xlabel('Speech index')
    axs[2].grid(True)

    plt.tight_layout()
    plt.show()
    
    # Scatter plot of v2 vs v3
    fig, ax = plt.subplots(figsize=(10,10))
    ax.scatter(v2, v3)

    for i in range(len(v2)):
        ax.annotate(years[i], (v2[i], v3[i]), fontsize=9)

    ax.set_xlabel('v2 (unit)')
    ax.set_ylabel('v3 (unit)')
    ax.set_title('Spectral Projection (Unit-Length tf-idf Vectors)')

    plt.show()
```

The structure here is dramatically different from that of the original un-normalized tf-idf vectors. It seems that since this normalization removes differences in overall magnitude (which might be results of simple document length or total word frequency), the new analysis is driven purely by the direction of each tf-idf vector. This means that this analysis is now more equivalent to cosine similarity in that it measures the pattern of term usage more closely. 

We also observe new shapes, directions, and clusters that could be corresponding to more subtle differences or progressions. For instance, in the first h value's scatter plot, there is a clear temporal trend, with more recent speeches more towards the left side and older speeches clusterd more towards the right. This could perhaps reflect shifts in topics or rhetoric over time -- this may not have been as clear in the original vectors where overall volume differences could have masked it.
